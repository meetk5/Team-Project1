```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "Data/allEmp_FoodServices&DrinkingPlaces.csv"
foodind_table = pd.read_csv(filepath)
```

```python
x_axis = foodind_table.loc[:,'Label']
y_axisfood = foodind_table.loc[:,'Value']
```

```python
plt.figure(figsize=(20,8))
barchart = plt.bar(x_axis,y_axisfood)
plt.xticks(rotation = 45)
plt.title ("Employment Trend in Food & Drink Services Industry in US")
plt.ylabel ("All Employees in Thousands")
barchart
```

```python
plt.figure(figsize=(20,8))
Lineplot = plt.plot(x_axis,y_axisfood)
plt.xticks(rotation = 45)
plt.title ("Employment Trend in Food & Drink Services Industry in US")
plt.ylabel ("All Employees in Thousands")
Lineplot
```
